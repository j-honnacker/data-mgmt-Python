```python
import pandas as pd
```

```python
tbl_a = pd.DataFrame({'store': ['Store A', 'Store B', 'Store C'],
                      '2017' : [   452000,   1204000,    870000],
                      '2018' : [   523000,   1450000,    910000],
                      '2019' : [   420000,   1020000,    810000]})
```

## `.melt()`

```python
tbl_a.melt(value_vars = ["2017",
                         "2018",
                         "2019"],
           var_name   = "year",
           value_name = "sales"
          )
```

```python
tbl_a.melt(id_vars    = "store",
           var_name   = "year",
           value_name = "sales"
          )
```

```python
# keep only sales for 2018 and 2019
tbl_a.melt(id_vars    = "store",
           value_vars = ["2017",
                         "2018"],
           var_name   = "year",
           value_name = "sales")
```
